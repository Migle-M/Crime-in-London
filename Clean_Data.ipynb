```python
import pandas as pd
import os
from datetime import datetime

crimedata1 = pd.read_csv("CrimeData18-19.csv")
crimedata2 = pd.read_csv("CrimeData19-20.csv")
```

```python
# join two datasets
crimedata = pd.concat([crimedata1, crimedata2], ignore_index = True)

# delete unneccessary columns
crimedata = crimedata.drop(columns=['Area name', 'Area code', 'Offence Subgroup', 'Measure', 'Financial Year'])

# convert date column type from object to datetime
crimedata['Month_Year']=pd.to_datetime(crimedata['Month_Year'])

# delete rows with unnecessary values
crimedata = crimedata.drop(crimedata[crimedata['Area Type'] == 'Safer Neighbourhood Teams'].index)
crimedata = crimedata.drop(columns=['Area Type'])
crimedata = crimedata.drop(crimedata[crimedata['Borough_SNT'] == 'Aviation Security(SO18)'].index)

# filter data to leave the data containing year 2019
crimedata = crimedata.drop(crimedata[crimedata['Month_Year'].dt.year == 2018].index)
crimedata = crimedata.drop(crimedata[crimedata['Month_Year'].dt.year == 2020].index)
crimedata.reset_index(drop=True, inplace=True)

crimedata.head()
```

```python
# sum the counts of offences for the whole year based on the borough and crime type
crime_type = crimedata.groupby(['Borough_SNT', 'Offence Group'])['Count'].sum().reset_index()
crime_type = crime_type.rename(columns={'Borough_SNT': 'Borough', 'Offence Group': 'Crime Type'})

crime_type.head()
```

```python
# export the dataframe as a csv file for the use for further data analysis and visualization
os.makedirs('data/', exist_ok=True)
crime_type.to_csv('data/crime_type.csv')
```

```python
# sum the counts of offences for the whole year based on the borough
crime_all = crime_type.groupby(['Borough'])['Count'].sum().reset_index()
crime_all.head()
```

```python
# export the dataframe as a csv file for the use for further data analysis and visualization
crime_all.to_csv('data/crime_all.csv')
```
